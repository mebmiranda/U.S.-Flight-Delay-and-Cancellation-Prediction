### Cleaning 2022 Data

```python
# Set the path to the folder where the CSV files are stored
folder_path = "/Users/maynardemmanuelmiranda/Desktop/CDA Project/Data/Data/2022"

# Get a list of all the CSV files in the folder
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Create an empty list to store the dataframes
dfs = []

# Loop through each CSV file and append the dataframe to the list
for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    dfs.append(df)

# Concatenate all the dataframes into one
combined_df_2022 = pd.concat(dfs, ignore_index=True)

# Print the first 5 rows of the combined dataframe
combined_df_2022.head()
```

```python
# select columns that do not start with "DIV"
filtered_df = combined_df_2022.filter(regex='^(?!DIV)')

# create a new DataFrame with only the selected columns
combined_df_2022 = pd.DataFrame(filtered_df)
```

```python
# Convert the 'FL_DATE' column to a datetime data type
combined_df_2022['FL_DATE'] = pd.to_datetime(combined_df_2022['FL_DATE'])

# Extract the date part only and overwrite the 'FL_DATE' column
combined_df_2022['FL_DATE'] = combined_df_2022['FL_DATE'].dt.date

# Print the first 5 rows of the modified DataFrame to verify the changes
combined_df_2022.head()
```

```python
# Not needed columns
cols_to_drop = ['OP_UNIQUE_CARRIER', 'OP_CARRIER_AIRLINE_ID', 'TAIL_NUM',
                'ORIGIN_AIRPORT_SEQ_ID', 'ORIGIN_CITY_MARKET_ID', 'ORIGIN_CITY_NAME',
                'ORIGIN_STATE_FIPS', 'ORIGIN_STATE_NM', 'ORIGIN_WAC',
                'DEST_AIRPORT_SEQ_ID', 'DEST_CITY_MARKET_ID', 'DEST_CITY_NAME',
                'DEST_STATE_NM', 'DEST_WAC']

# Drop the columns from the DataFrame
combined_df_2022 = combined_df_2022.drop(columns=cols_to_drop)
```

```python
# Top 10 busiest airports
valid_values = ['LAX', 'ORD', 'DEN', 'ATL', 'BOS', 'SFO', 'SEA', 'EWR', 'DFW', 'MCO']

# Filter the DataFrame to only include rows with valid values in the 'ORIGIN' and 'DEST' columns
combined_df_2022 = combined_df_2022[combined_df_2022['ORIGIN'].isin(valid_values) & combined_df_2022['DEST'].isin(valid_values)]

# Reset the index of the filtered DataFrame
combined_df_2022 = combined_df_2022.reset_index(drop=True)
#combined_df_2022 = combined_df_2022.dropna(axis=0)

# Print the first 5 rows of the filtered DataFrame to verify the changes
combined_df_2022.head()
```

```python
# convert "DEP_TIME" and "ARR_TIME" columns to datetime objects
df_final["DEP_TIME"] = pd.to_datetime(df_final["DEP_TIME"], format="%H:%M")
df_final["ARR_TIME"] = pd.to_datetime(df_final["ARR_TIME"], format="%H:%M")

# create a list of time range bins for the cut function
time_bins = [pd.to_datetime("00:00", format="%H:%M"), pd.to_datetime("05:59", format="%H:%M"), pd.to_datetime("11:59", format="%H:%M"), pd.to_datetime("17:59", format="%H:%M"), pd.to_datetime("23:59", format="%H:%M")]

# create a list of labels for the time range bins
time_labels = ["Early morning", "Morning", "Afternoon", "Evening"]

# create new columns "DEP_TIME_GROUP" and "ARR_TIME_GROUP"
df_final["DEP_TIME_GROUP"] = pd.cut(df_final["DEP_TIME"], bins=time_bins, labels=time_labels, include_lowest=True)
df_final["ARR_TIME_GROUP"] = pd.cut(df_final["ARR_TIME"], bins=time_bins, labels=time_labels, include_lowest=True)

df_final.head()
```
